### Line Plot

```python
plt.figure(dpi=150) 

# create a plot figure 
x = np.linspace(0, 10)

# create the first of two panels and set current axis 
plt.subplot(2, 1, 1) 

# (rows, columns, panel number) 
plt.plot(x, np.sin(x)) 

# create the second panel and set current axis 
plt.subplot(2, 1, 2) 
plt.plot(x, np.cos(x))
```

```python
# Extract the columns to be plotted
x = df.index
y1 = df['Glucose']
y3 = df['BMI']

plt.figure(figsize=[15, 5])

# Create the line plots
plt.plot(x, y1, label='Glucose')
plt.plot(x, y3, label='BMI')

# Add axis labels and a legend
plt.xlabel('Patient')
plt.ylabel('Value')
plt.legend()

# Display the plot
plt.show()
```

### Object oriented interface

```python
fig, ax = plt.subplots(2, dpi=150) 
# Call plot() method on the appropriate object 
ax[0].plot(x, np.sin(x)) 
ax[1].plot(x, np.cos(x))
```

```python
# Create a Figure and Axes object
fig, ax = plt.subplots()



# Set the x-axis and y-axis data
x = df.index
# y1 = df['Glucose']
y2 = df['BloodPressure']
y3 = df['BMI']

# Add the line plots to the Axes object
# ax.plot(x, y1, label='Glucose')
ax.plot(x, y2, label='BloodPressure')
ax.plot(x, y3, label='BMI')

# plt.figure(figsize=[15, 5]) - doen't work for the axes object

# Set the axis labels and legend
ax.set_xlabel('Patient')
ax.set_ylabel('Value')
ax.legend()

# Display the plot
plt.show()
```

### Scatter plots

```python
rng = np.random.RandomState(0) 

x = rng.randn(100) 
y = rng.randn(100) 

colors = rng.rand(100) 
sizes = 1000 * rng.rand(100) 

plt.figure(dpi=150) 
plt.xlim([-3, 3]) 
plt.ylim([-3, 3]) 
plt.scatter(x, y, c=colors, s=sizes, alpha=0.3) 

plt.colorbar()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
diabetes = pd.read_csv("diabetes.csv")

# Create a scatter plot using the 'plot' method of the DataFrame
diabetes.plot(kind='scatter', x='BMI', y='Glucose', c='Outcome', cmap='coolwarm', figsize=(8,6)) #figsize - modify the size

# Set the axis labels
plt.xlabel('BMI')
plt.ylabel('Glucose')

# Display the plot
plt.show()
```

### Error bars

```python
import random

number_of_points = 70 
x = np.linspace(0, 10, number_of_points) 
dy = 0.25 

np.random.seed(17) 

y = np.sin(x) + dy * np.random.randn(number_of_points) 

plt.figure(dpi=150) 
plt.errorbar(x, y, yerr = dy, fmt = '.k')
```

```python
# Compute the mean and standard deviation of the Glucose values for each Outcome group
glucose_mean = df.groupby('Outcome')['Glucose'].mean()
glucose_std = df.groupby('Outcome')['Glucose'].std()

# Set up the plot
fig, ax = plt.subplots()

# Add the error bars to the plot
ax.errorbar(x=[0,1], y=glucose_mean, yerr=glucose_std, fmt='o', capsize=5)

# Set the x-ticks and labels
ax.set_xticks([0,1])
ax.set_xticklabels(['No Diabetes', 'Diabetes'])

# Set the axis labels
ax.set_xlabel('Outcome')
ax.set_ylabel('Glucose')

# Display the plot
plt.show()
```

### Contour plots

```python
plt.style.use('seaborn-white') 

def f(x, y): 
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x) 

x = np.linspace(0, 5, 50) 
y = np.linspace(0, 5, 40) 

X, Y = np.meshgrid(x, y) 

Z = f(X,Y) 
plt.figure(dpi=150) 

plt.contour(X, Y, Z, 20, cmap='RdGy') 

plt.colorbar()
```

```python
from scipy.stats import gaussian_kde

# Set up the plot
fig, ax = plt.subplots()

# Create a meshgrid of x and y values
x = df['Age']
y = df['BMI']
X, Y = np.meshgrid(np.linspace(x.min(), x.max(), 100), np.linspace(y.min(), y.max(), 100))

# Compute the probability density function using a kernel density estimator
xy = np.vstack([x,y])
Z = gaussian_kde(xy)(np.vstack([X.ravel(), Y.ravel()]))
Z = Z.reshape(X.shape)

# Add the contour plot to the plot
ax.contour(X, Y, Z)

# Set the axis labels
ax.set_xlabel('Age')
ax.set_ylabel('BMI')

# Display the plot
plt.show()
```

### Histograms

```python
np.random.seed(0) 

x1 = np.random.normal(0, 0.8, 1000) 
x2 = np.random.normal(-2, 1, 1000) 
x3 = np.random.normal(4, 1, 1000) 

plt.figure(dpi=150) 
plt.hist(x1, alpha=0.3, bins=40) 
plt.hist(x2, alpha=0.3, bins=40) 
plt.hist(x3, alpha=0.3, bins=40)
```

```python
# Set up the plot
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Create histograms for each variable
axs[0, 0].hist(df['Pregnancies'], bins=10)
axs[0, 0].set_title('Pregnancies')

axs[0, 1].hist(df['Glucose'], bins=10)
axs[0, 1].set_title('Glucose')

axs[1, 0].hist(df['BloodPressure'], bins=10)
axs[1, 0].set_title('Blood Pressure')

axs[1, 1].hist(df['BMI'], bins=10)
axs[1, 1].set_title('BMI')

# Display the plot
plt.show()
```

### Bidimensional histogram

```python
mean = [0, 0] 
cov = [[1, 1], [1, 2]] 

np.random.seed(0) 
x, y = np.random.multivariate_normal(mean, cov, 10000).T 

plt.figure(dpi=150) 
plt.hist2d(x, y, bins=30, cmap='Blues') 
cb = plt.colorbar() 
cb.set_label('counts in bin')
```

```python
# Create the histogram
plt.hist2d(df['Glucose'], df['BMI'], bins=(20, 20), cmap='Blues')

# Set the title and axis labels
plt.title('Glucose vs. BMI')
plt.xlabel('Glucose')
plt.ylabel('BMI')

# Add a colorbar
plt.colorbar()

# Display the plot
plt.show()
```

### KDE plots

```python
with sns.axes_style('whitegrid'):
    plt.figure(dpi=150)
    
    # Create the KDE plot
    sns.kdeplot(diabetes['Glucose'], fill=True, label = 'Glucose')
    sns.kdeplot(diabetes['BMI'], fill=True, label = 'BMI')

    
    # Set the title and axis labels
    plt.title('Glucose Distribution')
    plt.xlabel('Glucose')
    plt.ylabel('Density')

    # display the legend
    plt.legend()

    # Display the plot
    plt.show()
```

### Bidimensional KDE plot

```python
with sns.axes_style('darkgrid'):
    # Create the KDE plot
    sns.kdeplot(data = df, x = df['Glucose'], y = df['BMI'], fill=True)
    
    # Set the title and axis labels
    plt.title('Glucose vs BMI')
    plt.xlabel('Glucose')
    plt.ylabel('BMI')
    
    # Display the plot
    plt.show()
```

### KDE jointplot

```python
# Create the jointplot
sns.jointplot(x = df['Glucose'], y = df['BMI'], kind='kde')

# Set the title and axis labels
plt.subplots_adjust(top = 0.95)
plt.suptitle('Glucose vs BMI', fontsize=14)
plt.xlabel('Glucose')
plt.ylabel('BMI')

# Display the plot
plt.show()
```
