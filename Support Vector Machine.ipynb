### Firstly, we will explore how the feature scale influences the algorithmâ€™s performance. Train an SVM model on 4 versions of the Data Set:

```python
# import the libraries
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler, MinMaxScaler, MaxAbsScaler
from sklearn.model_selection import train_test_split
```

```python
# split the dataset into features (X) and target (y)
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
# split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

__- Initial one.__

```python
# train and evaluate an SVM model on the initial dataset
svm = SVC(kernel='linear', C=1)

svm.fit(X_train, y_train)

initialScore = svm.score(X_test, y_test)
print("Accuracy on initial dataset:", initialScore)
```

__- The one passed through the StandardScaler.__

```python
# train and evaluate an SVM model on the dataset passed through the StandardScaler
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

svm.fit(X_train_scaled, y_train)

StandardScalerScore = svm.score(X_test_scaled, y_test)
print("Accuracy on StandardScaler dataset:", StandardScalerScore)
```

__- The one passed through the MinMaxScaler.__

```python
# train and evaluate an SVM model on the dataset passed through the MinMaxScaler
scaler = MinMaxScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

svm.fit(X_train_scaled, y_train)

MinMaxScalerScore = svm.score(X_test_scaled, y_test)
print("Accuracy on MinMaxScaler dataset:", MinMaxScalerScore)
```

__- The one passed through the MaxAbsScaler.__

```python
# train and evaluate an SVM model on the dataset passed through the MaxAbsScaler
scaler = MaxAbsScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

svm.fit(X_train_scaled, y_train)

MaxAbsScaler = svm.score(X_test_scaled, y_test)
print("Accuracy on MaxAbsScaler dataset:", MaxAbsScaler)
```

### Usually, SVMs use some Kernel to move the data to another space where linear planes separate the classes. In this task, you will have to explore how different kernels influence the decision boundary. Create the plots of decision boundary for the following kernels:

- linear
- poly
- rbf
- sigmoid 

```python
from sklearn import svm
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt

kernels = ['linear', 'poly', 'rbf', 'sigmoid']

fig, axs = plt.subplots(2, 2, figsize=(10, 10))
axs = axs.ravel()

for i, kernel in enumerate(kernels):
    clf = svm.SVC(kernel=kernel, C=1)
    clf.fit(X, y)

    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                         np.linspace(y_min, y_max, 100))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    axs[i].contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)
    axs[i].scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm)
    axs[i].set_title(kernel)

plt.suptitle("Decision Boundaries for different kernels")
plt.show()
```

### Plot them all together in a combined plot, compare them, and express some conclusions.

```python
plt.figure(figsize=(10, 10))

for kernel in kernels:
    clf = svm.SVC(kernel=kernel, C=1)
    clf.fit(X, y)

    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                         np.linspace(y_min, y_max, 100))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)

plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm)
plt.suptitle("Decision Boundaries for different kernels")
plt.legend(kernels)
plt.show()
```
